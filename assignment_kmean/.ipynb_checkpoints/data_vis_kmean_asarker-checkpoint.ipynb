```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Load and clean up automobile data 

```python

cols= ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style",
    "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type",
      "num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
        "peak-rpm","city-mpg","highway-mpg","price"]

car_data = pd.read_csv('imports-85.data', header=None,names=cols)
car_data.shape
```

```python
car_data.replace("?",np.nan,inplace=True)
car_data.dropna(inplace=True)
car_data.reset_index(drop=True, inplace=True)
car_data.head()
```

## Find out 6 continuous numerical attributes of  the Automobile Data Set

```python
car_data.corr()
```

```python
features = ["wheel-base", "length", "width", "height", "curb-weight", "engine-size"]
data =car_data[features]
sns.pairplot(data.dropna())
print('Pair plot of all variables')
```

# PCA Projection to 2D

```python
from sklearn.preprocessing import StandardScaler

# Separating out the features
x = car_data.loc[:, features].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
pcaDF = pd.DataFrame(data = principalComponents
             , columns = ['PCA 1', 'PCA 2'])
pcaDF.head()
```

# Visualize 2D Projection

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PCA 1', fontsize = 15)
ax.set_ylabel('PCA 2', fontsize = 15)
ax.set_title('PCA for Automobile data', fontsize = 20)
ax.scatter(pcaDF['PCA 1'], pcaDF['PCA 2'])
ax.grid()
plt.show()
```
